```python
import zipfile

import pandas as pd

import os

import glob
```

```python

names = [os.path.basename(x) for x in glob.glob('Flight_stats/*.zip')]

names
```

```python


appended_data = []

for i in names:

    filepath = (f"Flight_stats/{i}")
    
    zf = zipfile.ZipFile(filepath, allowZip64 = True)

    this = pd.read_csv(filepath)
    
    appended_data.append(this)
    
    this.to_csv()
    
    

    
    
   
    
    
    
    
    
    
```

```python
appended_data = pd.concat(appended_data)   

flight_frame = pd.DataFrame(appended_data)

flight_frame.to_csv('Flight_stats/all_flights_2016.csv')
```
